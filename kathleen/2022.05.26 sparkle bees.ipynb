```python
import numpy as np
import cv2
#vid = '../videos/smallbees.mov'
vid = '../videos/bees30sec.mov'
cap = cv2.VideoCapture(vid)

frames = []
for i in range(30):
    if i % 10 == 0:
        print(i)
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2. COLOR_BGR2GRAY).astype('float32')
    frames.append(gray)
    #frames.append(frame)
    
frames = np.array(frames)
frames.shape, frames.dtype
```

```python
medianFrame = np.median(frames, axis=0).astype(dtype=np.uint8)   
imshow(medianFrame, cmap='gray')
```

```python
beeFrame = frames[0]
figure(figsize=(20,20))
imshow(beeFrame, cmap='gray')
```

```python
foo = abs(medianFrame - beeFrame)
figure(figsize=(20,20))
imshow(foo)
```

```python
thing = abs(medianFrame - frames).max(0)
figure(figsize=(20,20))
imshow(thing)
```

```python
matplotlib.image.imsave('sparklebees.png', thing)
```
